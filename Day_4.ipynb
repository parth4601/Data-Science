### Separate the Inpot and Output variables

```python
x = df[['disp','hp','wt']]
y = df['mpg']
```

```python
x
```

### Import the class

```python
from sklearn.linear_model import LinearRegression
```

### Create the obj

```python
reg = LinearRegression()
```

### train the model

```python
reg.fit(x,y)
```

### Predict on unseen data

```python
disp = 221
hp =102
wt = 2.91
reg.predict([[disp,hp,wt]])
```

### visualize

```python
reg.score(x,y)
```

### Decision Tree Classification

### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df
```

### separate the input and output variables

```python
x = df[['Age','EstimatedSalary']]
y = df['Purchased']
```

```python
x
```

### explore the data

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'])
```

```python
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c=y)
```

### import the class 

```python
from sklearn.tree import DecisionTreeClassifier
```

### create the obj

```python
classifier = DecisionTreeClassifier(random_state=0)
```

### train the model

```python
classifier.fit(x,y)
```

### plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,15))
plot_tree(classifier ,fontsize=7 , feature_names=['age','sal'],
           class_names=['y','n'], filled=True ,rounded=True);
```

### Prediction on unseen data

```python
new = pd.DataFrame({
    'Age':[25,37,56,66],
    'EstimatedSalary':[34000,41000,26000,134000]
})
```

```python
new
```

```python
classifier.predict(new)
```

### Naive bayes Classification

```python
from sklearn.naive_bayes import GaussianNB
```

```python
# train
classifier.fit(x,y)
```
